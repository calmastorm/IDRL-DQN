### different task density

```python
episodes=[100,200,500,1000,2000]

episodes_mean_rewards=np.load("episodes_mean_rewards.npy")

font_legend = {'family' : 'Arial','weight' : 'normal','size': 20}

font_title = {'family' : 'Arial','weight' : 'normal','size': 20}

font_label = {'family' : 'Arial','weight' : 'normal','size': 25}

fig, axs = plt.subplots(figsize = (20,10))

labels = axs.get_xticklabels() + axs.get_yticklabels()

[label.set_fontsize(20) for label in labels]

[label.set_fontname('Times New Roman') for label in labels]

axs.grid()

axs.plot(episodes,episodes_mean_rewards, linestyle='--',linewidth=4,color='red',marker='x',markersize=20)


xmajorLocator   = MultipleLocator(100)
   
axs.xaxis.set_major_locator(xmajorLocator)

axs.set_xlabel('Total episodes',font_label)

axs.set_ylabel('Mean reward',font_label)



plt.show()
```

```python
buffer_sizes=[1000,5000,10000,20000,40000]

buffer_size_rewards=np.load('buffer_size_rewards.npy')

font_legend = {'family' : 'Arial','weight' : 'normal','size': 20}

font_title = {'family' : 'Arial','weight' : 'normal','size': 20}

font_label = {'family' : 'Arial','weight' : 'normal','size': 25}

fig, axs = plt.subplots(figsize = (20,10))

labels = axs.get_xticklabels() + axs.get_yticklabels()

[label.set_fontsize(20) for label in labels]

[label.set_fontname('Times New Roman') for label in labels]

axs.grid()

axs.plot(buffer_sizes,buffer_size_rewards, linestyle='--',linewidth=4,color='blue',marker='x',markersize=20)

xmajorLocator   = MultipleLocator(5000)
   
axs.xaxis.set_major_locator(xmajorLocator)

axs.set_xlabel('Buffer sizes',font_label)

axs.set_ylabel('Mean reward',font_label)


plt.show()
```
